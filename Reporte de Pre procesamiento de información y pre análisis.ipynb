# Reporte de Pre procesamiento de información y pre análisis


> 1) Análisis de datos

> 2) Consultoría con creadores de contenido

> 3) desarrollo e implementación de inteligencia artificial personalizada a las necesidades del proyecto


Posterior al desarrollo de la el análisis del estado del arte, el desarrollo de herramientas para capturar y limpiar la base de datos, la manufactura de herramientas de visualización estadística para el análisis de la distribución de las bases de datos se realizó un diagnóstico del proyecto para determinar cuál es el plan de acción para desarrollar  modelos de inteligencia artificial eficientes que tengan el mayor impacto para el creador de contenido.
el diagnóstico consistió en 3 fases:


#### 1) Análisis de datos
El primer paso para el diagnóstico fue el análisis de datos de la muestra seleccionada, la cual se encuentra en el presente documento y fueron generadas las siguientes conclusiones; Si bien fueron  encontradas muchas correlaciones positivas para el incremento visualizaciones para el creador de contenido, una simple  interpretación estadística del canal no es suficiente para  una mejora significativa en su rendimiento. Esto se debe a un conjunto de factores, El primero y más importante es que nuestra base de datos en la actualidad es estática y está fija en el tiempo, esto no nos permite generar 3 resultados estadísticos prioritarios para determinar lo saludable que es un canal dentro de la plataforma, este es un problema actual que será resuelto en el momento que se tengan usuarios que utilicen nuestros servicios de una forma recurrente y tendremos la posibilidad de generar metadata relacionada en tiempo. Esto permitirá poder tomar bases de datos divididas en ‘épocas temporales’ para poder determinar 2 factores de alto impacto:


* a) CTR: el ctr o (click through rate) determina el incremento o disminución de frecuencia con la cual se reproduce un video, esta es una pieza de información altamente valiosa para el éxito de un creador de contenido en la plataforma, ya que puede ser un termómetro que determine a corto plazo si los cambios en el contenido o tu tendencia es favorable o desfavorable para la cantidad de visualizaciones. Esto le ayudará a tomar decisiones efectivas en periodos de tiempo muy cortos.

* b) Churn rate: esta es la relación de abandono de la página del creador del contenido y es una contraparte al CTR. Esta métrica te permite obtener el porcentaje de abandono de tu web en función de los seguidores, la cantidad total en base al total de tus suscriptores, o las pérdidas que ha generado ese abandono de visitas a la web.


* c) Un tercer marcador de importancia es de índole cualitativa, la retroalimentación de los suscriptores del canal.Una opinión global determinada por palabras clave unidas a sentimientos puede mejorar la comunicación entre el creador de contenido y el suscriptor. Esta es una tarea en la que no son requeridas bases de datos adicionales, los comentarios ya existen.


#### 2) Consultoría con creadores de contenido:
El segundo paso fue comunicarnos con creadores de contenido para determinar cuáles son las mayores necesidades que buscan satisfacer. La retroalimentación fue muy clara, es muy evidente que la necesidad que tienen actualmente los creadores de contenido es poder determinar su ctr y su churn rate, sin embargo también se hizo mención a la necesidad de desarrollo de herramientas para la clasificación y generación de thumbnails es una necesidad.


Los thumbnails son el primer contacto que tiene el usuario de la plataforma hacia el contenido, gran parte de la industria de youtube enfoca la mayor cantidad de esfuerzos en la generación de thumbnails , muchas veces generando hasta 10 thumbnails diferentes para un solo video, ya que un thumbnail atractivo mejora automáticamente el ctr.


Esto nos dio la idea de buscar bases de datos de thumbnails para desarrollar modelos predictivos para la clasificación del tipo de contenido. Se obtuvo una base de  2515 imágenes de thumbnails divididos en 24 diferentes categorías para modelación de redes neuronales.

Dentro de esta retroalimentación fue también muy evidente la necesidad de una herramienta que pueda determinar el ánimo de la audiencia de una forma eficaz a través de los comentarios. El creador de contenido no tiene el tiempo ni el conocimiento como para tomar la muestra de comentarios de su video y poder determinar en promedio cuáles son las emociones dominantes de la audiencia.

tomando en cuenta esto y contemplando la base de datos que ya hemos generado decidimos implementar un NLU capaz de determinar emociones dentro de la web app para los comentarios de un video, para esto fue necesario generar un algoritmo a través del desarrollo de nuestro NLP



#### 3)  desarrollo e implementación de inteligencia artificial personalizada a las necesidades del proyecto
Finalmente el tercer paso para el desarrollo diagnóstico y toma de decisiones fue, una vez determinados los problemas actuales con soluciones factibles, nos dimos a la tarea de desarrollar las herramientas de software que nos permitan crear modelos predictivos para el análisis de sentimientos de los comentarios, generar una red neuronal convolucional para generar algoritmos entrenados de clasificación visual de thumbnails  y desarrollar la arquitectura del software para el soporte de nuestras aplicaciones en una página web.


La toma de decisiones fue desarrollada posterior al tomar en cuenta todos estos aspectos  y el contenido en las bases de datos que ya hemos generado decidimos implementar un NLU capaz de determinar emociones dentro de la web app para los comentarios de un video, para esto fue necesario generar un algoritmo a través del desarrollo de nuestro NLP
