Posterior al desarrollo de la el anlisis del estado del arte, el desarrollo de herramientas para capturar y limpiar la base de datos, la manufactura de herramientas de visualizacin estadstica para el anlisis de la distribucin de las bases de datos se realiz un diagnstico del proyecto para determinar cul es el plan de accin para desarrollar  modelos de inteligencia artificial eficientes que tengan el mayor impacto para el creador de contenido.
el diagnstico consisti en 3 fases:


#### 1) Anlisis de datos
El primer paso para el diagnstico fue el anlisis de datos de la muestra seleccionada, la cual se encuentra en el presente documento y fueron generadas las siguientes conclusiones; Si bien fueron  encontradas muchas correlaciones positivas para el incremento visualizaciones para el creador de contenido, una simple  interpretacin estadstica del canal no es suficiente para  una mejora significativa en su rendimiento. Esto se debe a un conjunto de factores, El primero y ms importante es que nuestra base de datos en la actualidad es esttica y est fija en el tiempo, esto no nos permite generar 3 resultados estadsticos prioritarios para determinar lo saludable que es un canal dentro de la plataforma, este es un problema actual que ser resuelto en el momento que se tengan usuarios que utilicen nuestros servicios de una forma recurrente y tendremos la posibilidad de generar metadata relacionada en tiempo. Esto permitir poder tomar bases de datos divididas en pocas temporales para poder determinar 2 factores de alto impacto:


* a) CTR: el ctr o (click through rate) determina el incremento o disminucin de frecuencia con la cual se reproduce un video, esta es una pieza de informacin altamente valiosa para el xito de un creador de contenido en la plataforma, ya que puede ser un termmetro que determine a corto plazo si los cambios en el contenido o tu tendencia es favorable o desfavorable para la cantidad de visualizaciones. Esto le ayudar a tomar decisiones efectivas en periodos de tiempo muy cortos.

* b) Churn rate: esta es la relacin de abandono de la pgina del creador del contenido y es una contraparte al CTR. Esta mtrica te permite obtener el porcentaje de abandono de tu web en funcin de los seguidores, la cantidad total en base al total de tus suscriptores, o las prdidas que ha generado ese abandono de visitas a la web.


* c) Un tercer marcador de importancia es de ndole cualitativa, la retroalimentacin de los suscriptores del canal.Una opinin global determinada por palabras clave unidas a sentimientos puede mejorar la comunicacin entre el creador de contenido y el suscriptor. Esta es una tarea en la que no son requeridas bases de datos adicionales, los comentarios ya existen.


#### 2) Consultora con creadores de contenido:
El segundo paso fue comunicarnos con creadores de contenido para determinar cules son las mayores necesidades que buscan satisfacer. La retroalimentacin fue muy clara, es muy evidente que la necesidad que tienen actualmente los creadores de contenido es poder determinar su ctr y su churn rate, sin embargo tambin se hizo mencin a la necesidad de desarrollo de herramientas para la clasificacin y generacin de thumbnails es una necesidad.


Los thumbnails son el primer contacto que tiene el usuario de la plataforma hacia el contenido, gran parte de la industria de youtube enfoca la mayor cantidad de esfuerzos en la generacin de thumbnails , muchas veces generando hasta 10 thumbnails diferentes para un solo video, ya que un thumbnail atractivo mejora automticamente el ctr.


Esto nos dio la idea de buscar bases de datos de thumbnails para desarrollar modelos predictivos para la clasificacin del tipo de contenido. Se obtuvo una base de  2515 imgenes de thumbnails divididos en 24 diferentes categoras para modelacin de redes neuronales.

Dentro de esta retroalimentacin fue tambin muy evidente la necesidad de una herramienta que pueda determinar el nimo de la audiencia de una forma eficaz a travs de los comentarios. El creador de contenido no tiene el tiempo ni el conocimiento como para tomar la muestra de comentarios de su video y poder determinar en promedio cules son las emociones dominantes de la audiencia.

tomando en cuenta esto y contemplando la base de datos que ya hemos generado decidimos implementar un NLU capaz de determinar emociones dentro de la web app para los comentarios de un video, para esto fue necesario generar un algoritmo a travs del desarrollo de nuestro NLP



#### 3)  desarrollo e implementacin de inteligencia artificial personalizada a las necesidades del proyecto
Finalmente el tercer paso para el desarrollo diagnstico y toma de decisiones fue, una vez determinados los problemas actuales con soluciones factibles, nos dimos a la tarea de desarrollar las herramientas de software que nos permitan crear modelos predictivos para el anlisis de sentimientos de los comentarios, generar una red neuronal convolucional para generar algoritmos entrenados de clasificacin visual de thumbnails  y desarrollar la arquitectura del software para el soporte de nuestras aplicaciones en una pgina web.


La toma de decisiones fue desarrollada posterior al tomar en cuenta todos estos aspectos  y el contenido en las bases de datos que ya hemos generado decidimos implementar un NLU capaz de determinar emociones dentro de la web app para los comentarios de un video, para esto fue necesario generar un algoritmo a travs del desarrollo de nuestro NLP
